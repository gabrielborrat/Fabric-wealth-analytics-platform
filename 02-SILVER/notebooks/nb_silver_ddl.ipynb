```python
%%sql
CREATE TABLE IF NOT EXISTS silver_ctl_entity (
  entity_code           STRING,   -- ex: transactions, fx_rates, mcc, users, cards
  enabled               BOOLEAN,
  load_order            INT,
  notebook_name         STRING,    -- ex: nb_silver_transactions
  notebook_path         STRING,    -- ex: /Lakehouse/Notebooks/nb_silver_transactions (ou votre path workspace)
  load_mode_default     STRING,    -- full|incremental
  partition_cols        ARRAY<STRING>,  -- ex: ["txn_month"] / [] si non partitionn√©
  natural_key_cols      ARRAY<STRING>,  -- ex: ["transaction_id"] ou ["base_currency","currency","fx_date"]
  depends_on            ARRAY<STRING>,  -- ex: ["users","cards","mcc","fx_rates"] (facultatif si load_order suffit)
  critical              BOOLEAN,   -- si true : fail pipeline si KO
  timeout_minutes       INT,       -- ex: 60
  retries               INT,       -- ex: 1
  owner                 STRING,    -- optionnel
  notes                 STRING,    -- optionnel
  updated_ts            TIMESTAMP
)
USING delta;

```

```python
%%sql
CREATE TABLE IF NOT EXISTS silver_log_runs (
  run_id           STRING,
  pipeline_name    STRING,
  layer            STRING,      -- 'silver'
  start_ts         TIMESTAMP,
  end_ts           TIMESTAMP,
  status           STRING,      -- RUNNING|SUCCESS|FAILED|PARTIAL
  triggered_by     STRING,      -- manual|schedule|ci
  environment      STRING,      -- dev|prod
  params_json      STRING,
  error_message    STRING
)
USING delta;
```
