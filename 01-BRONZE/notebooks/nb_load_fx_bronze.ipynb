```sql
%%pyspark
# PARAMETERS (cellule à marquer comme paramètre)
LandingPath = "Files/landing/fx/"       # valeur par défaut pour les tests
BronzeTableName = "bronze_fx_raw"       # valeur par défaut pour les tests
```

```sql
%%pyspark
from pyspark.sql import functions as F

# Lecture du landing
df_fx_raw = (
    spark.read
         .option("header", True)
         .csv(LandingPath)
)

# Transformations Bronze
df_fx = (
    df_fx_raw
      .withColumn("currency", F.upper(F.col("currency")))
      .withColumn("date", F.to_date("date"))
      .withColumn("rate_vs_usd", F.col("exchange_rate").cast("double"))
      .select(
          "currency",
          "base_currency",
          "currency_name",
          "rate_vs_usd",
          "date"
      )
)

# Écriture Bronze
(df_fx
    .write
    .mode("append")
    .saveAsTable(BronzeTableName)
)





```
